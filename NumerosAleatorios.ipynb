##Practica 1. Generacion de 100 numeros aleatorios por el metodo de centros al cuadrado y por el metodo de medios cuadrados

```python
#metodo de centros cuadrados
# semilla inicial
semilla = 2004
n = 4
iteraciones = 100
resultados = []

for _ in range(iteraciones):
  #elevar la semilla al cuadrado
  cuadrado = str(semilla ** 2).zfill(2 * n )# agregar los esdpacios necesarios
  # extraer los digitos centrales
  inicio = (len (cuadrado) - n) // 2
  semilla = int (cuadrado[inicio:inicio + n])
  resultados.append(semilla)

  #imprimir resultados
  for i, numero in enumerate(resultados, 1):
    print(f"Numero aleatrorio,{i}:{numero}")
```

```python
#metodo de medios cuadrados
# Semilla inicial
semilla = 5731
n = 4
iteraciones = 100
numeros_generados = []

for _ in range(iteraciones):
    # Elevar la semilla al cuadrado y formatear la longitud
    cuadrado = str(semilla ** 2).zfill(2 * n)

    # Extraer los primeros y últimos n/2 dígitos
    primera_parte = cuadrado[:n//2]  # Primeros n/2 dígitos
    segunda_parte = cuadrado[-(n//2):]  # Últimos n/2 dígitos
    semilla = int(primera_parte + segunda_parte)  # Concatenar para formar la nueva semilla

    numeros_generados.append(semilla)

# Imprimir los resultados
for indice, valor in enumerate(numeros_generados, 1):
    print(f"Número aleatorio {indice}: {valor}")
```
